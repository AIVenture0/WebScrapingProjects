```python
from urllib.request import Request,urlopen
from requests import get
import pandas as pd
from bs4 import BeautifulSoup
Title=[]
Time=[]
Genre=[]
Certificate=[]
Rating=[]
Votes=[]
Metascore=[]
Collection=[]
url = 'http://www.imdb.com/search/title?release_date=2017&sort=num_votes,desc&page=1'
req=Request(url,headers={'User-Agent': 'Mozilla/5.0'})
response=urlopen(req)


html_soup = BeautifulSoup(response, 'html.parser')
type(html_soup)
movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')

for data in movie_containers:
    vin=data.find_all('h3',{'class':'lister-item-header'})
    for i in vin:
        vik=i.find('a').text
        
        Title.append(vik)
        
for data1 in movie_containers:
    new=data1.find_all('span',{'class':'genre'})
    for i in new:
        Genre.append(i.find(text=True))
        
    new1=data1.find_all('span',{'class':'runtime'})
    for i in new1:
        Time.append(i.find(text=True))
        
        
    new2=data1.find_all('span',{'class':'certificate'})
    for iv in new2:
        vik=iv.find(text=True)
        Certificate.append(vik)
        
    new3=data1.find_all('strong')
    for i in new3:
        Rating.append(float(i.find(text=True)))
        
        
    new4= data1.find_all('span',{'class':'metascore favorable'})
    for i in new4:
        Metascore.append(int(i.find(text=True)))
        
    
    new5=data1.find('span',attrs={'name':'nv'})
    #Votes.append(new5.find(text=True)))
    votes=new5.find(text=True)
    ne=int(votes.replace(',',''))
    Votes.append(ne)
    #here is we use find_all then it the resule also come up with the gross collection but it is not here so try to find out how 
    # to seperapte out it too.
    # form the given data.
    
    for vin in data1:
        
        try:
            new=vin.find_all('span',attrs={'name':'nv'})[1]
            
            Collection.append(new.find(text=True))
            break
                
        except:
            pass
        
#here we try to collect the collection of the movies,
#the collection and the votes are at the same 'span'tag we try to collect the data form both 
#their is A GOOD concept while collecting this data out .
    
    
   
   

    
    
movi_data= %time pd.DataFrame(Title,columns=['Title'])
    
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
class ScrapSIC(object):
    def __call__(self):
        page = urlopen('http://www.sec.gov/info/edgar/siccodes.htm')
        soup = bs(page.read(),'lxml')
        table = soup.find_all('table')[2]
        table = table.find('table')
        trs = table.find_all('tr')[3:]
        for tr in trs:
            tds = tr.find_all('td')
            sic_code = tds[0].string
            sic_name = tds[-1].string.title()
            print(sic_code, sic_name)
s=%time ScrapSIC()
```
